```python
import os
import json
import email.utils
path = os.path.abspath('main.ipynb')
with open(os.path.join(os.path.dirname(path), "datajson/table.json"), 'r', encoding='utf8') as f:
    table = json.loads(f.read()) # Таблица решений задач
    
with open(os.path.join(os.path.dirname(path), "datajson/failed.json"), 'r', encoding='utf8') as f:
    failed = json.loads(f.read()) # Данные по ошибкам
    
with open(os.path.join(os.path.dirname(path), "datajson/messages.json"), 'r', encoding='utf8') as f:
    message = json.loads(f.read()) # Полученные сообщения
    
messages = [(m['subj'].upper().split(" ")[:-1], email.utils.parsedate(m['date'])) for m in message]
days = [(m['subj'].upper(), m['date'].split(',')) for m in message]
print(table['data'][0][0].upper())
data = [(m[0], m[3]) for m in table['data']]
datatask = [(m[2], m[3]) for m in table['data']]
print(datatask)
```

```python
import matplotlib.pyplot as plt
import numpy as np
xtime = [i for i in range(0,24)]
ytime = [0 for i in range(0,24)]
tmp = [tmp[1][3] for tmp in messages]
for y0 in tmp:
    ytime[y0] = ytime[y0]+1
fig, ax = plt.subplots()
ax.bar(xtime, ytime)
ax.set_xlabel('Время (ч)')
ax.set_ylabel('Количество сообщений')
fig.set_figwidth(12)
fig.set_figheight(6)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xday = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
yday = [0 for i in range(0, 7)]
tmp = [tmp[1][0] for tmp in days]
for y0 in tmp:
  yday[xday.index(y0)] = yday[xday.index(y0)]+1
fig, ax = plt.subplots()
ax.bar(xday, yday)
ax.set_xlabel('День недели')
ax.set_ylabel('Количество сообщений')
fig.set_figwidth(12)
fig.set_figheight(6)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xgroup = []
ygroup = []
for tmp in messages:
    if not tmp[0][0] in xgroup:
        xgroup.append(tmp[0][0])
        ygroup.append(1)
    else:
        ygroup[xgroup.index(tmp[0][0])] += 1
fig, ax = plt.subplots()
ax.bar(xgroup, ygroup)
ax.set_xlabel('Группа')
ax.set_ylabel('Количество сообщений')
fig.set_figwidth(20)
fig.set_figheight(6)

plt.show()
```
